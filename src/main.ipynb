## Parte 1: Entrenamiento Incremental

### 1.1. Investigación Teórica

El entrenamiento incremental, una técnica que actualiza los modelos de manera continua a medida que llegan nuevos datos, ha ganado atención significativa en el campo del aprendizaje automático y profundo. En esta sección, nos sumergiremos en una revisión bibliográfica para comprender las capacidades y limitaciones del entrenamiento incremental en dos algoritmos seleccionados: Random Forest y Redes Neuronales.

Random Forest es un algoritmo de aprendizaje supervisado que utiliza un conjunto de árboles de decisión para realizar predicciones (Brieman, 2001). Una de las características clave de Random Forest es su capacidad para manejar grandes conjuntos de datos con múltiples características, lo que lo hace atractivo para aplicaciones de entrenamiento incremental (Ge, He & Zeng, 2017). Sin embargo, el proceso de actualización de un Random Forest existente con nuevos datos puede ser computacionalmente costoso, ya que implica la reevaluación de múltiples árboles y la combinación de sus resultados.

En términos de entrenamiento incremental, se ha investigado la adaptación de Random Forest para actualizar de manera eficiente sus estimaciones a medida que llegan nuevos datos (Gama, 2010). Algunos enfoques se centran en la actualización de los pesos de los árboles existentes, mientras que otros exploran estrategias para agregar nuevos árboles al bosque existente de manera incremental. Estos esfuerzos buscan mejorar la eficiencia computacional del proceso de actualización, sin comprometer la precisión del modelo.

Por otro lado, las Redes Neuronales son modelos computacionales inspirados en el funcionamiento del cerebro humano, que se utilizan ampliamente en una variedad de aplicaciones de aprendizaje automático y profundo (LeCun, Bengio & Hinton, 2015). Una de las ventajas clave de las Redes Neuronales es su capacidad para capturar relaciones complejas y no lineales en los datos, lo que las hace especialmente adecuadas para problemas de alta dimensionalidad como la detección de fraudes en transacciones financieras (Goodataellow, Bengio & Courville, 2016).

En el contexto del entrenamiento incremental, las Redes Neuronales pueden actualizarse con nuevos datos utilizando técnicas como el descenso de gradiente estocástico y el aprendizaje en línea (Hinton, Osindero & Teh, 2006). Estos enfoques permiten que las redes neuronales se adapten de manera continua a medida que llegan nuevos datos, sin la necesidad de volver a entrenar el modelo desde cero. Sin embargo, el entrenamiento incremental de Redes Neuronales puede plantear desafíos en términos de la gestión de la memoria y la estabilidad del modelo, especialmente en entornos de datos cambiantes y ruidosos (Bottou, 2010).

Mientras tanto, la columna "unix_time" nos brinda la oportunidad de calcular la duración entre transacciones o identificar tendencias temporales más complejas.

En cuanto a las variables numéricas presentes en el dataset, como el monto de la transacción ("amt"), la latitud y longitud ("lat", "long"), la población de la ciudad ("city_pop"), y otras relacionadas con la cantidad y frecuencia de transacciones, exploramos diversas posibilidades de derivar nuevas características. Para las variables categóricas como el comerciante ("merchant"), la categoría de transacción ("category"), el género ("gender") y el trabajo ("job"), implementamos técnicas de codificación adecuadas para transformarlas en características numéricas que los modelos de aprendizaje automático puedan utilizar. Finalmente, consideramos características derivadas a partir de columnas existentes, como la proporción de transacciones fraudulentas por usuario o la identificación de usuarios que realizan su primera transacción en un comercio específico. Estas características adicionales ayudan a capturar patrones más sutiles y complejos en los datos, mejorando así la capacidad de los modelos para detectar fraudes de manera más precisa y efectiva. En conjunto, la ingeniería de características juega un papel crucial en la mejora de la representación de los datos y, en última instancia, en el rendimiento de los modelos de detección de fraudes.

El enfoque que se tomará en cuenta en este procedimiento de ingeniería de características con las características que se encuentran presentes en este dataset consiste en ir de columna en columna obteniendo información y características importantes y útiles, y finalizar con la creación de características generales que necesiten el uso de múltiples columnas. La primera columna que se encuentra en el dataset que nos fue proporcionado es "trans_date_trans_time", que incluye la fecha y hora en la que fue realizada la transacción.

```python
# Valores de la columna trans_date_trans_time.
data["trans_date_trans_time"]
```

Podemos observar cómo la columna tiene un objeto de fecha muy específico, que contiene el año, mes, día, hora, minuto y segundo de la transacción. Esta columna, como se puede observar también, es tratada como un objeto y no como una fecha ni un atributo numérico de ningún tipo. Por esta razón, esta columna puede ser transformada a varias columnas mucho más útiles consistentes de entradas numéricas que representen cada uno de los diferentes datos individuales e independientes que se ubican en la columna. Para poder extraer las características útiles que podemos utilizar, necesitamos una función que podamos aplicar a la columna actual múltiples veces con el objetivo de extraer estas características. La función es llamada `split_date_and_time`, y nos devuelve un objeto con los seis valores útiles que podemos extraer de la columna.

```python
# Función split_date_and_time, que retorna todos los valores de una fecha.
def split_date_and_time(entry):

    # Diccionario que almacenará el resultado obtenido del proceso.
    result = dict()

    # Conversión de la fecha a string.
    entry = str(entry)

    # Obtención de la fecha y de la hora, separando por espacios.
    date, time = entry.split(" ")

    # Obtención del año, mes y día, separando la fecha por guiones.
    year, month, day = date.split("-")

    # Almacenamiento del año, mes y día obtenidos.
    result["year"] = int(year)
    result["month"] = int(month)
    result["day"] = int(day)

    # Obtención de la hora, minuto y segundo, separando la hora por dos puntos.
    hour, minute, second = time.split(":")

    # Almacenamiento de la hora, minuto y segundo obtenidos.
    result["hour"] = int(hour)
    result["minute"] = int(minute)
    result["second"] = int(second)

    # Retorno del resultado final.
    return result
```

Con la función anterior desarrollada, podemos probarla con una fecha aleatoria que se encuentre dentro del dataset, con el objetivo de verificar que las separaciones de datos y el retorno del objeto necesitado sean realizados de manera correcta y la función pueda utilizarse.

```python
# Prueba de la función desarrollada en una entrada.
split_date_and_time(data["trans_date_trans_time"].iloc[0])
```

Todo está funcionando como debería. La función nos retorna un objeto con los seis valores que necesitamos para crear las columnas que queremos utilizar. Ahora sólo necesitamos crear cada una de las columnas aplicando la función a "trans_date_trans_time" con la función `apply`, y obtener estas seis columnas que resultarán mucho más útiles para el entrenamiento del modelo que únicamente manejar la fecha y tener qué convertirla a una columna categórica que no nos diría nada de la transacción. A continuación se aplica la función a la columna mencionada seis veces, para crear las seis columnas necesarias.

```python
# Obtención del año de la transacción.
data["transaction_year"] = data["trans_date_trans_time"].apply(lambda entry: split_date_and_time(entry).get("year"))
data["transaction_year"]
```

```python
# Obtención del mes de la transacción.
data["transaction_month"] = data["trans_date_trans_time"].apply(lambda entry: split_date_and_time(entry).get("month"))
data["transaction_month"]
```

```python
# Obtención del día de la transacción.
data["transaction_day"] = data["trans_date_trans_time"].apply(lambda entry: split_date_and_time(entry).get("day"))
data["transaction_day"]
```

```python
# Obtención de la hora de la transacción.
data["transaction_hour"] = data["trans_date_trans_time"].apply(lambda entry: split_date_and_time(entry).get("hour"))
data["transaction_hour"]
```

```python
# Obtención del minuto de la transacción.
data["transaction_minute"] = data["trans_date_trans_time"].apply(lambda entry: split_date_and_time(entry).get("minute"))
data["transaction_minute"]
```

```python
# Obtención del segundo de la transacción.
data["transaction_second"] = data["trans_date_trans_time"].apply(lambda entry: split_date_and_time(entry).get("second"))
data["transaction_second"]
```

```python
import pandas as pd
import numpy as np
from datetime import datetime
from geopy.distance import geodesic

# Supongamos que tu DataFrame se llama data
# Asegúrate de tener las librerías necesarias instaladas: pandas, numpy, geopy

# Función para calcular la distancia entre dos coordenadas
def calculate_distance(lat1, long1, lat2, long2):
    return geodesic((lat1, long1), (lat2, long2)).kilometers

# Función para obtener el día de la semana
def get_day_of_week(year, month, day):
    return datetime(year, month, day).weekday()

# Crear nuevas columnas
data["transaction_time_of_day"] = pd.cut(data["transaction_hour"], bins=[0, 6, 12, 18, 24], labels=["Madrugada", "Mañana", "Tarde", "Noche"], right=False)
data["transaction_day_of_week"] = data.apply(lambda row: get_day_of_week(row["transaction_year"], row["transaction_month"], row["transaction_day"]), axis=1)
data["transaction_is_weekend"] = data["transaction_day_of_week"].isin([5, 6]).astype(int)

# Distancia entre el cliente y el comerciante
data["distance_client_merchant"] = data.apply(lambda row: calculate_distance(row["transaction_lat"], row["transaction_long"], row["transaction_merch_lat"], row["transaction_merch_long"]), axis=1)

# Logaritmo del monto de la transacción
data["log_transaction_amount"] = np.log1p(data["transaction_amount"])

# Porcentaje del gasto mensual/anual
data["transaction_percent_month"] = data["transaction_amount"] / data["transaction_amount_month"]
data["transaction_percent_year"] = data["transaction_amount"] / data["transaction_amount_year"]

# Frecuencia de uso de la tarjeta
data["transaction_freq_cc"] = data.groupby("transaction_cc_num_first_four_digits")["transaction_cc_num_first_four_digits"].transform("count")

# Diversidad de comerciantes
data["transaction_merchant_diversity"] = data.groupby("transaction_cc_num_first_four_digits")["transaction_merchant"].transform("nunique")

# Patrones de gasto por género y profesión
gender_avg_spend = data.groupby("transaction_gender")["transaction_amount"].transform("mean")
data["gender_avg_spend_diff"] = data["transaction_amount"] - gender_avg_spend

job_avg_spend = data.groupby("transaction_job")["transaction_amount"].transform("mean")
data["job_avg_spend_diff"] = data["transaction_amount"] - job_avg_spend

# Frecuencia de transacciones en periodos de tiempo
data["transaction_count_last_day"] = data.groupby("transaction_cc_num_first_four_digits")["transaction_unix_time"].transform(lambda x: x.diff().fillna(0).lt(86400).sum())
data["transaction_count_last_week"] = data.groupby("transaction_cc_num_first_four_digits")["transaction_unix_time"].transform(lambda x: x.diff().fillna(0).lt(604800).sum())
data["transaction_count_last_month"] = data.groupby("transaction_cc_num_first_four_digits")["transaction_unix_time"].transform(lambda x: x.diff().fillna(0).lt(2592000).sum())

# Diversidad de categorías de transacciones
data["transaction_category_diversity"] = data.groupby("transaction_cc_num_first_four_digits")["transaction_category"].transform("nunique")

# Interacciones entre monto y hora del día
data["amount_time_of_day"] = data["transaction_amount"] * data["transaction_hour"]

# Interacciones entre distancia y monto
data["amount_distance"] = data["transaction_amount"] * data["distance_client_merchant"]

# Finalmente, visualizar las primeras filas del DataFrame para verificar los nuevos features
data.head()
```

```python
import pandas as pd
import numpy as np
from datetime import datetime
from geopy.distance import geodesic

# Suponiendo que el DataFrame original se llama data

# Feature 1: Hora del día (periodos del día)
def get_time_period(hour):
    if 5 <= hour < 12:
        return "morning"
    elif 12 <= hour < 17:
        return "afternoon"
    elif 17 <= hour < 21:
        return "evening"
    else:
        return "night"

data["transaction_time_period"] = data["transaction_hour"].apply(get_time_period)

# # Feature 2: Día de la semana
# data["transaction_date"] = pd.to_datetime(data[["transaction_year", "transaction_month", "transaction_day"]])
# data["transaction_day_of_week"] = data["transaction_date"].dt.dayofweek

# Feature 3: Fin de semana
data["transaction_is_weekend"] = data["transaction_day_of_week"].apply(lambda x: 1 if x >= 5 else 0)

# Feature 4: Estacionalidad
def get_season(month):
    if month in [12, 1, 2]:
        return "winter"
    elif month in [3, 4, 5]:
        return "spring"
    elif month in [6, 7, 8]:
        return "summer"
    else:
        return "fall"

data["transaction_season"] = data["transaction_month"].apply(get_season)

# Feature 5: Distancia entre cliente y comerciante
def calculate_distance(row):
    coords_1 = (row["transaction_lat"], row["transaction_long"])
    coords_2 = (row["transaction_merch_lat"], row["transaction_merch_long"])
    return geodesic(coords_1, coords_2).kilometers

data["distance_customer_merchant"] = data.apply(calculate_distance, axis=1)

# Feature 6: Transacción redondeada
# Ya está presente como `transaction_round_amount`

# Feature 7: Logaritmo del monto
data["transaction_amount_log"] = np.log(data["transaction_amount"] + 1)  # Agregar 1 para evitar log(0)

# Feature 8: Porcentaje del gasto mensual/anual
data["transaction_percentage_month"] = data["transaction_amount"] / (data["transaction_amount_month"] + 1e-5)
data["transaction_percentage_year"] = data["transaction_amount"] / (data["transaction_amount_year"] + 1e-5)

# Feature 9: Desviación estándar del monto de la transacción
data["transaction_amount_std_month"] = data.groupby("transaction_month")["transaction_amount"].transform("std")
data["transaction_amount_std_year"] = data.groupby("transaction_year")["transaction_amount"].transform("std")

# Feature 10: Frecuencia de uso de la tarjeta
data["transaction_count_by_card"] = data.groupby("transaction_cc_num_first_four_digits")["transaction_cc_num_first_four_digits"].transform("count")

# Feature 11: Diversidad de comerciantes
data["unique_merchants_by_card"] = data.groupby("transaction_cc_num_first_four_digits")["transaction_merchant"].transform("nunique")

# Feature 12: Patrones de gasto por género y profesión
data["avg_amount_by_gender"] = data.groupby("transaction_gender")["transaction_amount"].transform("mean")
data["avg_amount_by_job"] = data.groupby("transaction_job")["transaction_amount"].transform("mean")

# Feature 13: Historial de transacciones
# data["transaction_count_last_day"] = data.groupby("transaction_date")["transaction_amount"].transform("count")
# data["transaction_count_last_week"] = data.groupby(pd.Grouper(key="transaction_date", freq="W"))["transaction_amount"].transform("count")
# data["transaction_count_last_month"] = data.groupby(pd.Grouper(key="transaction_date", freq="M"))["transaction_amount"].transform("count")

# Feature 14: Variedad de categorías de transacciones
data["unique_categories_by_card"] = data.groupby("transaction_cc_num_first_four_digits")["transaction_category"].transform("nunique")

# Feature 15: Interacciones entre monto y hora del día
data["amount_time_period_interaction"] = data["transaction_amount"] * data["transaction_hour"]

# Feature 16: Interacciones entre distancia y monto
data["amount_distance_interaction"] = data["transaction_amount"] * data["distance_customer_merchant"]

def haversine(lat1, lon1, lat2, lon2):
    """
    Calculate the great-circle distance between two points on the Earth surface.
    Input coordinates are in decimal degrees.
    """
    # Convertir grados a radianes
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    # Diferencias de coordenadas
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    # Fórmula de Haversine
    a = np.sin(dlat / 2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    
    # Radio de la Tierra en kilómetros (6371)
    km = 6371 * c
    return km

# Aplicar la función Haversine al DataFrame
data["distance_client_merchant"] = data.apply(lambda row: haversine(row["transaction_lat"], row["transaction_long"], row["transaction_merch_lat"], row["transaction_merch_long"]), axis=1)

# Ahora puedes revisar las nuevas columnas añadidas
data.head()
```

```python
# Transacciones por hora del día
data["transaction_count_per_hour"] = data.groupby("transaction_hour")["transaction_hour"].transform("count")

# Transacciones por día de la semana
data["transaction_count_per_weekday"] = data.groupby("transaction_day_of_week")["transaction_day_of_week"].transform("count")

# Monto total por comerciante
data["total_amount_per_merchant"] = data.groupby("transaction_merchant")["transaction_amount"].transform("sum")

# Monto promedio por comerciante
data["avg_amount_per_merchant"] = data.groupby("transaction_merchant")["transaction_amount"].transform("mean")

# Número de transacciones por comerciante
data["transaction_count_per_merchant"] = data.groupby("transaction_merchant")["transaction_merchant"].transform("count")

# Tiempo desde la última transacción de la misma tarjeta
data = data.sort_values(by=["transaction_cc_num_first_four_digits", "transaction_unix_time"])
data["time_since_last_transaction"] = data.groupby("transaction_cc_num_first_four_digits")["transaction_unix_time"].diff()

# Tiempo desde la primera transacción del día para la misma tarjeta
data["time_since_first_transaction_of_day"] = data.groupby(["transaction_cc_num_first_four_digits", "transaction_year", "transaction_month", "transaction_day"])["transaction_unix_time"].transform(lambda x: x - x.min())

# Transacciones en la misma ciudad
data["transaction_count_same_city"] = data.groupby(["transaction_cc_num_first_four_digits", "transaction_city_pop"])["transaction_city_pop"].transform("count")

# Cambio en el monto de la transacción
data["transaction_amount_change"] = data.groupby("transaction_cc_num_first_four_digits")["transaction_amount"].diff()

# Cambio en la ubicación
data["transaction_lat_change"] = data.groupby("transaction_cc_num_first_four_digits")["transaction_lat"].diff()
data["transaction_long_change"] = data.groupby("transaction_cc_num_first_four_digits")["transaction_long"].diff()
data["location_change"] = data.apply(lambda row: haversine(row["transaction_lat"], row["transaction_long"], row["transaction_lat_change"] + row["transaction_lat"], row["transaction_long_change"] + row["transaction_long"]) if not pd.isna(row["transaction_lat_change"]) and not pd.isna(row["transaction_long_change"]) else 0, axis=1)

# Frecuencia de uso de la tarjeta en diferentes categorías
data["transaction_count_per_category"] = data.groupby(["transaction_cc_num_first_four_digits", "transaction_category"])["transaction_category"].transform("count")

# Monto promedio de transacciones por categoría
data["avg_amount_per_category"] = data.groupby(["transaction_cc_num_first_four_digits", "transaction_category"])["transaction_amount"].transform("mean")

# Monto acumulado en los últimos 7 días
data["transaction_amount_last_7_days"] = data.groupby("transaction_cc_num_first_four_digits")["transaction_amount"].rolling(window=7, min_periods=1).sum().reset_index(level=0, drop=True)

# Monto acumulado en el último mes
data["transaction_amount_last_month"] = data.groupby("transaction_cc_num_first_four_digits")["transaction_amount"].rolling(window=30, min_periods=1).sum().reset_index(level=0, drop=True)

# Desviación estándar del monto de las transacciones
data["std_transaction_amount"] = data.groupby("transaction_cc_num_first_four_digits")["transaction_amount"].transform("std")

# Coeficiente de variación del monto de las transacciones
data["cv_transaction_amount"] = data["std_transaction_amount"] / data.groupby("transaction_cc_num_first_four_digits")["transaction_amount"].transform("mean")

# Visualizar las primeras filas del DataFrame para verificar los nuevos features
data.head()
```
